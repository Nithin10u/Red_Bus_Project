```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time
import pandas as pd

# Set up the WebDriver
driver = webdriver.Chrome()
driver.maximize_window()

#state route page hyperlink
driver.get("https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile")

time.sleep(3)

wait = WebDriverWait(driver, 5)

def scroll_page(n):
    SCROLL_PAUSE_TIME = n

    last_height = driver.execute_script("return document.body.scrollHeight")

    while True:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

        time.sleep(SCROLL_PAUSE_TIME)

        new_height = driver.execute_script("return document.body.scrollHeight")
        if new_height == last_height:
            break
        last_height = new_height

scroll_page(5)
bus_route_name_list = []
bus_route_link_list = []

page_number = 1

for i in range(0,3):

    try:
        bus_route_name = driver.find_elements(By.CSS_SELECTOR,"a[class='route']")
        for bus in bus_route_name:    
            bus_route_name_list.append(bus.text)
            bus_route_link_list.append(bus.get_attribute("href"))

        pagination_container = driver.find_element(By.XPATH,'//*[@id="root"]/div/div[4]/div[12]')
        next_page_button = pagination_container.find_element(By.XPATH,f'.//div[contains(@class,"DC_117_pageTabs") and text() = "{int(page_number) + 1}"]')
        
        actions = ActionChains(driver)
        actions.move_to_element(next_page_button).perform()
        wait.until(EC.element_to_be_clickable(next_page_button))
        next_page_button.click()
        page_number += 1

        time.sleep(2)

    except:
        print("No such page") 

df= pd.DataFrame({"bus_route_name":bus_route_name_list,"bus_route_link":bus_route_link_list})   
       
```

```python
df.to_csv("AndhraPradesh_Data.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
import pandas as pd
import time

# --------- Chrome setup ----------
options = Options()
options.add_argument("--start-maximized")
options.add_argument("--disable-popup-blocking")
options.add_argument("--disable-notifications")

# ✅ Update chromedriver path here
service = Service("C:/path/to/chromedriver.exe")
driver = webdriver.Chrome()

# --------- Read route CSV ----------
df = pd.read_csv("AndhraPradesh_Data.csv", usecols=["bus_route_name", "bus_route_link"])

# --------- Lists to store data ----------
bus_route_list = []
bus_link_list = []
bus_name_list = []
bus_type_list = []
departing_time_list = []
duration_list = []
reaching_time_list = []
seat_available_list = []
star_rating_list = []
price_list = []

# --------- Function to wait for bus list ----------
def wait_for_buses():
    for _ in range(15):
        buses = driver.find_elements(By.CSS_SELECTOR, "li[class*='tupleWrapper']")
        if len(buses) > 0:
            return buses
        time.sleep(1)
    return []

# --------- Scraping loop ----------
for idx, link in enumerate(df["bus_route_link"]):
    print(f"Scraping route {idx+1}/{len(df)}: {link}")
    driver.get(link)
    time.sleep(5)  # allow React to load

    # Scroll to load all buses (lazy load)
    for i in range(5):
        driver.execute_script("window.scrollBy(0, 1000);")
        time.sleep(1.5)

    buses = wait_for_buses()
    print(f"Found {len(buses)} buses for this route.")

    for bus in buses:
        bus_route_list.append(df["bus_route_name"][idx])
        bus_link_list.append(link)

        # ✅ Updated Selectors based on your snippet

        # Bus Name
        try:
            bus_name = bus.find_element(By.CSS_SELECTOR, ".travelsName___d34747").text.strip()
        except:
            bus_name = ""
        bus_name_list.append(bus_name)

        # Bus Type
        try:
            bus_type = bus.find_element(By.CSS_SELECTOR, ".busType___a2b374").text.strip()
        except:
            bus_type = ""
        bus_type_list.append(bus_type)

        # Departure Time
        try:
            depart_time = bus.find_element(By.CSS_SELECTOR, ".boardingTime___abc589").text.strip()
        except:
            depart_time = ""
        departing_time_list.append(depart_time)

        # Arrival Time
        try:
            arrival_time = bus.find_element(By.CSS_SELECTOR, ".droppingTime___162d68").text.strip()
        except:
            arrival_time = ""
        reaching_time_list.append(arrival_time)

        # Duration
        try:
            duration = bus.find_element(By.CSS_SELECTOR, ".duration___a35daf").text.strip()
        except:
            duration = ""
        duration_list.append(duration)

        # Seats Available
        try:
            seats = bus.find_element(By.CSS_SELECTOR, ".totalSeats___493dc2").text.strip()
        except:
            seats = ""
        seat_available_list.append(seats)

        # Star Rating
        try:
            rating = bus.find_element(By.CSS_SELECTOR, ".rating___ca6690").text.strip()
        except:
            rating = "0.0"
        star_rating_list.append(rating)

        # Price
        try:
            price = bus.find_element(By.CSS_SELECTOR, ".finalFare___634549").text.strip()
        except:
            price = ""
        price_list.append(price)

# --------- Save to CSV ----------
df_final = pd.DataFrame({
    "bus_route": bus_route_list,
    "bus_link": bus_link_list,
    "bus_name": bus_name_list,
    "bus_type": bus_type_list,
    "departing_time": departing_time_list,
    "duration": duration_list,
    "reaching_time": reaching_time_list,
    "seat_available": seat_available_list,
    "star_rating": star_rating_list,
    "price": price_list
})

df_final.to_csv("AndhraPradesh_Bus_Data.csv", index=False, encoding="utf-8-sig")
print("✅ Scraping completed successfully! Saved to AndhraPradesh_Bus_Data.csv")

driver.quit()
```

```python
import pandas as pd

# Read the original route CSV
df_routes = pd.read_csv("AndhraPradesh_Data.csv")

# Drop any unnamed/index columns automatically
df_routes = df_routes.loc[:, ~df_routes.columns.str.contains("^Unnamed")]

# Read the scraped bus data CSV
df_buses = pd.read_csv("AndhraPradesh_Bus_Data.csv")

# Drop any unnamed/index columns automatically
df_buses = df_buses.loc[:, ~df_buses.columns.str.contains("^Unnamed")]

# Rename route CSV columns to match bus data (if they exist)
if "bus_route_name" in df_routes.columns:
    df_routes = df_routes.rename(columns={"bus_route_name": "bus_route"})
if "bus_route_link" in df_routes.columns:
    df_routes = df_routes.rename(columns={"bus_route_link": "bus_link"})

# Merge on route and link columns, keep all buses
df_merged = pd.merge(df_buses, df_routes, on=["bus_route", "bus_link"], how="left")

# Remove duplicate columns if any
df_merged = df_merged.loc[:, ~df_merged.columns.duplicated()]

# Save to a single CSV
df_merged.to_csv("AndhraPradesh_All_Bus_Data.csv", index=False, encoding="utf-8-sig")
print("✅ Both CSVs safely merged into 'AndhraPradesh_All_Bus_Data.csv'")
```
